```python
covid19_1=covid19[['2차접종']].iloc[1:,1:].reset_index()
```

```python
global_distriction_jsonurl='https://raw.githubusercontent.com/southkorea/southkorea-maps/master/kostat/2018/json/skorea-provinces-2018-geo.json'
```

```python
global_dict = json.loads(requests.get(global_distriction_jsonurl).text)
```

```python
[global_dict['features'][i]['properties']['name'] for i in range(17)]
```

```python
second=pd.read_csv('https://raw.githubusercontent.com/guebin/2021DV/master/_notebooks/2021-11-22-prov.csv')
```

```python
second
```

```python
covid19_total=pd.concat([second,covid19_1],axis=1)
```

```python
covid19_total
```

```python
q=covid19_total.iloc[:,1]
```

```python
q
```

```python
w=covid19_total.iloc[:,3]
```

```python
covid19_total2=covid19_total.assign(prob=list(map(lambda x,y: x/y,covid19_total.iloc[:,3] ,covid19_total.iloc[:,1])))
```

```python
covid19_total2.iloc[:,1:]
```

```python
m = folium.Map([35.84195368311022, 127.1155556693179],zoom_start=6,scrollWheelZoom=False)
choro= folium.Choropleth(
    data=covid19_total2, 
    geo_data= global_dict, 
    color="prob",
    columns=['행정구역(시군구)별','prob'], 
    key_on='feature.properties.name'
)
choro.add_to(m) 
#m
```
