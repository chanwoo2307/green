```python
import pandas as pd
mpg = pd.read_csv("mpg.csv", header=0) # csv 파일 불러오기
```

```python
mpg.shape
```

```python
mpg.head(6) 
```

```python
mpg.info() #info함수를 사용하여 자료의 구조를 살펴보시오
```

```python
mpg.describe() #변수들에 대하여 데이터의 타입에 따른 요약통계량을 계산하시오.
```

```python
mpg.describe(include='all') #데이터 보기
```

```python
mpg.describe(include='object')
```

```python
mpg.dtypes #mpg 데이터 유형 확인
```

```python
mpg['manufacturer'].value_counts() #manufacturer 범주형 빈도분석
```

```python
mpg['model'].value_counts() #moder 범주형 빈도분석
```

```python
mpg['trans'].value_counts() #trans 범주형 빈도분석
```

```python
mpg['drv'].value_counts() #drv 범주형 빈도분석
```

```python
pd.crosstab(mpg['trans'],mpg['drv'], margins=True)# trans, drv별 교차분석
```

```python
pd.crosstab(mpg['trans'],mpg['drv']).apply(lambda r: r/len(mpg), axis=1) #전체 빈도 비율 확인 
```

```python
import pandas as pd
census = pd.read_csv("http://archive.ics.uci.edu/ml/machine-learning-databases/adult/adult.data", header=None)
census.columns = ['age', 'workclass', 'fnlwgt', 'eduction', 'education.num', 'marital.status', 
                   'occupation', 'relationship', 'race', 'sex', 'capital.gain', 'capital.loss', 
                   'hours.per.week', 'native.country', 'income']
```

```python
census.head()
```

```python
census.info() #info함수를 이용하여 자료의 구조를 살펴보시오.
```

```python
census.describe() #각 변수들에 대하여 데이터의 타입에 따른 요약통계량을 계산하시오.
```

```python
census.describe(include='all')
```

```python
census.describe(include='object')
```

```python
census
```

```python
census.dtypes #데이터 유형 확인
```

```python
census['workclass'].value_counts() #workclass 범주형 빈도분석
```

```python
census['eduction'].value_counts()  #education 범주형 빈도분석
```

```python
census['marital.status'].value_counts() #marital.status  범주형 빈도분석
```

```python
pd.crosstab(census['eduction'], census['marital.status']) #eduction, marital.status별 교차 분석
```

```python
pd.crosstab(census['eduction'], census['marital.status']).apply(lambda r: r/len(mpg), axis=1) #전체 빈도 비율 확인 
```
